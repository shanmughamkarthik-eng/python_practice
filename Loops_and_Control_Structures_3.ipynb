```

```

### Problem 4.6: Diamond Pattern (Hard)

Print a diamond pattern for n rows (n should be odd):

```
    *
   ***
  *****
 *******
*********
 *******
  *****
   ***
    *
```
(Example shown for n=9)

**Hint:** Print pyramid (upper half) for rows 1 to (n+1)//2, then print inverted pyramid (lower half) for remaining rows.

````python
print("Print a diamond pattern for n rows (n should be odd):")
my_number = int(input("Enter a number: "))

if my_number % 2 == 0:
    print("Please enter an odd number")

else:
    half = (my_number+1)//2
    print("")

# Top-Half pattern
    for i in range(1, half+1):
        for j in range(half-i):
            print(" ", end="")
        for j in range(2*i-1):
            print("*", end="")
        print("")

# Bottom-Half pattern
    for i in range(half-1, 0, -1):
        for j in range(half-i):
            print(" ", end="")
        for j in range(2*i-1):
            print("*", end="")
        print("")
````

### Problem 5.1: Find First Multiple (Easy)

Write a program that finds and prints the first number in the range 1 to 100 that is divisible by both 7 and 13. Use the `break` statement to exit the loop once found.

**Hint:** Loop through 1 to 100. Check if i%7==0 and i%13==0. If true, print i and use break to exit loop.

````python
print("Finding the first number in the range 1 to 100 that is divisible by both 7 and 13.")

for number in range(1, 101):
  if number % 7 == 0  and number % 13 == 0:
      print(f"The first number {number} is divisible by both 7 and 13. ")
      break
````

### Problem 5.2: Skip Multiples (Easy)

Write a program that prints all numbers from 1 to 50, but skips (doesn't print) numbers that are multiples of 3. Use the `continue` statement.

**Hint:** Loop from 1 to 50. If i%3==0, use continue to skip to next iteration. Otherwise print i.

````python
print("Write a program that prints all numbers from 1 to 50, but skips numbers that are multiples of 3")

for num in range(1, 51):
  if num % 3 == 0:
    continue
  else:
    print(num)
````

### Problem 5.3: Sum Until Negative (Medium)

Write a program that:
1. Keeps asking the user to enter numbers
2. Adds positive numbers to a sum
3. Skips zero (continues to next iteration)
4. Stops when a negative number is entered
5. Prints the final sum

**Hint:** Use while True loop. If num < 0, break. If num == 0, continue. Otherwise add to sum.

````python
print("Sum Until Negative")

sum = 0
Flag = True

while Flag:
    number = int(input(f"The current Sum is: {sum}.\t Enter a number: "))

    if number < 0:
        Flag = False
        break

    elif number == 0:
        continue

    else:
        sum += number

print(f"The final sum is : {sum}")
````
