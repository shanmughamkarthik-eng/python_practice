### Problem 6.1: Prime Numbers in Range (Medium)

Write a program to print all prime numbers between two given numbers (inclusive).

Example: Between 10 and 30: 11, 13, 17, 19, 23, 29

**Hint:** Outer loop for each number in range. For each number, use inner loop to check if it's prime (like Problem 2.5).

```python

print("Write a program to print all prime numbers between two given numbers (inclusive).")

my_number_start = int(input("Enter a number for starting range: "))
my_number_final = int(input("Enter a number for final range: "))

print(f"\nPrime numbers between {my_number_start} and {my_number_final} are: ")
for my_number in range(my_number_start, my_number_final + 1):
    if my_number <= 1:
        print(f"{my_number} is not a prime number")

    elif my_number == 2:
        print(f"{my_number} is a prime number")

    else:
        for i in range(2, my_number):
          if (my_number % i) == 0:
            #print(f"{my_number} is not a prime number")
            break
        else:
            print(my_number , end = " ")
```

### Problem 6.2: GCD Calculator (Medium)

Write a program to find the Greatest Common Divisor (GCD) of two numbers using the Euclidean algorithm.

Algorithm:
1. Divide the larger number by the smaller number
2. Replace the larger number with the smaller number
3. Replace the smaller number with the remainder
4. Repeat until the remainder is 0
5. The GCD is the last non-zero remainder

**Hint:** While b != 0: store remainder (a%b), then a=b, b=remainder. When loop ends, a is the GCD.

```python
print("Write a program to find the Greatest Common Divisor (GCD) of two numbers using the Euclidean algorithm")

number_a = int(input("Enter a number: "))
number_b = int(input("Enter a number: "))

while number_b:
      reminder = number_a % number_b
      number_a = number_b
      number_b = reminder

print(f"The GCD is : {number_a}")
```

### Problem 6.3: Perfect Number Checker (Hard)

Write a program to check if a number is a perfect number.

A perfect number is a positive integer that is equal to the sum of its proper positive divisors (excluding the number itself).

Example: 6 is perfect because 6 = 1 + 2 + 3

Example: 28 is perfect because 28 = 1 + 2 + 4 + 7 + 14

**Hint:** Loop from 1 to n-1. If i divides n (n%i==0), add i to sum. Compare sum with n.

```python
print("Write a program to check if a number is a perfect number.")

my_number = int(input("Enter a number: "))
total_sum = 0
sum_divisor = []

for number in range(1, my_number):
    if my_number % number == 0:
        total_sum = total_sum + number
        sum_divisor.append(number)

divisor_string = " + ".join(map(str, sum_divisor))

if total_sum == my_number:
    print(f"{my_number} is a perfect number because divisors {divisor_string} = {total_sum}")

else:
    print(f"{my_number} is not a perfect number")
```

### Problem 6.4: LCM Calculator (Medium)

Write a program to find the Least Common Multiple (LCM) of two numbers.

Hint: LCM(a, b) = (a Ã— b) / GCD(a, b)

You can use the GCD function from Problem 6.2.

**Hint:** First calculate GCD using Euclidean algorithm (Problem 6.2), then use formula: LCM = (a * b) // GCD.

```python
print("Write a program to find the Least Common Multiple (LCM) of two numbers.")

number_a = int(input("Enter a number: "))
number_b = int(input("Enter a number: "))

temp_a = number_a
temp_b = number_b

while temp_b:
  reminder = temp_a % temp_b
  temp_a = temp_b
  temp_b = reminder

lcm = (number_a * number_b) // temp_a
print(f"The LCM of {number_a} and {number_b} is : {lcm}")
```

### Problem 6.6: Strong Number Checker (Hard)

Write a program to check if a number is a Strong number.

A Strong number is a number whose sum of factorial of digits is equal to the original number.

Example: 145 is a strong number because 1! + 4! + 5! = 1 + 24 + 120 = 145

**Hint:** Extract each digit, calculate its factorial, add to sum. Compare sum with original number.

```python
print("Write a program to check if a number is a Strong number.")
my_number = int(input("Enter a number: "))
total_sum = 0
sum_divisor = []
temp_number = my_number

if my_number == 0:
    total_sum = 1
    sum_divisor = [1]
else:
    while temp_number > 0:
        reminder = temp_number % 10
        temp_number = temp_number // 10

        factorial = 1
        for number in range(1, reminder + 1):
            factorial *= number

        total_sum += factorial
        sum_divisor.append(factorial)

sum_divisor.reverse()
divisor_string = " + ".join(map(str, sum_divisor))

if total_sum == my_number:
    print(f"{my_number} is a strong number because {divisor_string} = {total_sum}")
else:
    print(f"{my_number} is not a strong number because {divisor_string} = {total_sum} (not {my_number})")
```
