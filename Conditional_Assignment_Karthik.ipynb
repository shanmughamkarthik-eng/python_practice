```python
# Exercise 1: Simple Age Check Write a program that checks if a person can get a driver's license (age >= 16).
print("\nWrite a program that checks if a person can get a driver's license (age >= 16)")
age = int(input("Please enter your age: "))
if age >= 16:
    print("You can get a driver's license.")
else:
    print("You cannot get a driver's license yet.")
print("-"*50)
```

```python
# Exercise 2: Positive or Negative Write a program that checks if a number is positive, negative, or zero.
print("\nWrite a program that checks if a number is positive, negative, or zero.")
number = float(input("Please enter a number: "))
if number >= 0:
    if number == 0:
      print("The number is zero.")
    else:
      print("The number is positive.")
else:
    print("The number is negative.")
print("-"*50)
```

```python
# Exercise 3: Larger Number Write a program that compares two numbers and prints which one is larger.
print("\nWrite a program that compares two numbers and prints which one is larger.")
number1 = float(input("Please enter the first number: "))
number2 = float(input("Please enter the second number: "))
if number1 == number2:
    print("The numbers are equal.")
else:
    if number1 > number2:
        print("The first number is larger.")
    else:
        print("The second number is larger.")
print("-"*50)
```

```python
# Exercise 4: Calculator Discount A store offers:
# 10% discount if purchase > $100
# 5% discount if purchase > $50
# No discount otherwise
# Calculate the final price.
print("\nExercise 4: Calculator Discount A store offers:")
print("10% discount if purchase > $100")
print("5% discount if purchase > $50")
print("No discount otherwise")
purchase = float(input("Please enter the purchase amount: "))
if purchase > 100:
    discount = 0.1
    final_price = purchase * (1 - discount)
elif purchase > 50:
    discount = 0.05
    final_price = purchase * (1 - discount)
else:
    discount = 0
    final_price = purchase

print(f"The final price after discount is: ${final_price:.2f}")
print("-"*50)
```

```python
# Exercise 5: Leap Year Checker Write a program to check if a year is a leap year:
# Divisible by 4 AND (not divisible by 100 OR divisible by 400)
print("\nExercise 5: Leap Year Checker Write a program to check if a year is a leap year:")
print("\nDivisible by 4 AND (not divisible by 100 OR divisible by 400)")
year = int(input("Please enter a year: "))
if (year % 4 == 0 and year % 100 != 0) or (year % 400 == 0):
    print(f"{year} is a leap year.")
else:
    print(f"{year} is not a leap year.")
print("-"*50)
```

```python
# Exercise 6: Grade Calculator with Bonus Calculate grades (A-F)
# but add +1 bonus grade level if attendance > 90%.
print("\nExercise 6: Grade Calculator with Bonus Calculate grades (A-F)")
print("\nadd +1 bonus grade level if attendance > 90%.")
grades = ["F", "D", "C", "B", "A"]
attendance = float(input("Enter attendance %: "))
current_grade = input("Enter current grade (A-F): ").upper()

if current_grade in grades:
    grade_index = grades.index(current_grade)
    if attendance > 90 and grade_index < 4:
        grade_index += 1

    final_grade = grades[grade_index]
    print(f"Final Grade: {final_grade} (Grade Points: {grade_index})")
else:
    print("Invalid grade entered.")
print("-"*50)
```

```python
# Exercise 7: Login System with Attempts Create a login system that:
# Checks username and password
# Tracks login attempts (max 3)
# Locks account after failed attempts
print("\nExercise 7: Login System with Attempts Create a login system that:")
print("\nChecks username and password")
print("Tracks login attempts (max 3)")
print("Locks account after failed attempts")
username = "admin"
password = "password123"
max_attempts = 3
attempts = 0

while attempts < max_attempts:
    input_username = input("Enter username: ")
    input_password = input("Enter password: ")

    if input_username == username and input_password == password:
        print("Login successful!")
        break

    else:
        attempts += 1
        print(f"Login failed. Attempts left: {max_attempts - attempts}")
        if attempts == max_attempts:
            print("Account locked. Maximum attempts reached.")
print("-"*50)
```

```python
# Exercise 8: Shipping Cost Calculator Calculate shipping based on:
# Weight of package
# Destination (domestic/international)
# Express or standard shipping
print("\nExercise 8: Shipping Cost Calculator Calculate shipping based on:")
print("\nWeight of package")
print("Destination (domestic/international)")
print("Express or standard shipping")
weight = float(input("Enter package weight (kg): "))
destination = input("Enter destination (domestic/international): ").lower()
shipping_type = input("Enter shipping type (express/standard): ").lower()

if destination == "domestic":
    rate_per_kg = 5.0                   # ₹5.00 per kg
else:
    rate_per_kg = 15.0                  # ₹15.00 per kg (International)

rate_per_type = 5.0                     # ₹5.00 per kg (Standard)
if shipping_type == "express":
    rate_per_type *= 1.5                # 50% extra for speed

total_cost = weight * rate_per_kg * rate_per_type

print(f"Total shipping cost: ${total_cost:.2f}")
```
