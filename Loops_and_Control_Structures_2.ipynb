### Problem 3.3: Number Guessing Game (Medium)

Write a number guessing game where:
1. The program generates a random number between 1 and 100
2. The user has to guess the number
3. After each guess, the program tells the user if their guess is too high or too low
4. The game continues until the user guesses correctly
5. Display the number of attempts taken

**Hint:** Use `import random` and `random.randint(1, 100)` to generate number. Use while loop with condition guess != target. Count attempts.

````python
print("Number Guessing Game")
import random
random_number = random.randint(1, 100)
count = 0

while True:
    my_number = int(input("Enter a number: "))
    count += 1
    if my_number > random_number:
        print("The guess is too high")

    elif my_number < random_number:
        print("The guess is too low")

    else:
        print(f"\nYou guessed it right. The random number is {my_number}")
        break

    print("Please, Try again.")

print(f"Number of attempts guessed: {count}")
````

### Problem 3.4: Palindrome Checker (Medium)

Write a program to check if a given number is a palindrome or not.

A palindrome number reads the same forwards and backwards.

Examples: 121, 12321, 1331 are palindromes

**Hint:** Store original number, reverse it (like Problem 3.1), then compare original with reversed number.

````python
print("Write a program to check if a given number is a palindrome or not.")

my_number = int(input("Enter a number: "))
temp = my_number
reverse = 0

while temp > 0:
    reverse = (reverse * 10) + (temp%10)
    temp = temp // 10

if my_number == reverse:
    print(f"The number {my_number } is a palindrome")
else:
    print(f"The number {my_number } is not a palindrome")
````

### Problem 3.5: Armstrong Number (Hard)

Write a program to check if a number is an Armstrong number (also called Narcissistic number).

An Armstrong number is a number that is equal to the sum of its own digits each raised to the power of the number of digits.

Examples:
- 153 = 1³ + 5³ + 3³ = 1 + 125 + 27 = 153 (Armstrong number)
- 9474 = 9⁴ + 4⁴ + 7⁴ + 4⁴ = 6561 + 256 + 2401 + 256 = 9474 (Armstrong number)

**Hint:** First count digits (power). Store original. Extract each digit, raise to power, add to sum. Compare sum with original.

````python
print("Write a program to check if a number is an Armstrong number.")

my_number = int(input("Enter a number: "))
length = len(str(my_number))
my_digit_list = []

temp = my_number
sum_digits = 0

while temp > 0:
    number = temp%10
    number_power = number ** length
    my_digit_list.append(number_power)
    sum_digits += number_power
    temp = temp // 10

my_digit_list.reverse()
#print(my_digit_list)

if sum_digits == my_number:
    print(f"Yes, {my_number} is an Armstrong number!")
else:
    print(f"No, {my_number} is not an Armstrong number.")
````

## Section 4: Nested Loops

### Problem 4.1: Pattern Printing - Right Triangle (Easy)

Print the following pattern for n rows:

```
*
**
***
****
*****
```
(Example shown for n=5)

**Hint:** Use nested loops: outer loop for rows (i from 1 to n), inner loop for columns (j from 1 to i). Print * in inner loop.

````python
print("Print the pattern:")
my_number = int(input("Enter a number: "))
for i in range(1, my_number+1):
    for j in range(1,i+1):
        print("*", end="")
    print("")
````

### Problem 4.2: Pattern Printing - Inverted Right Triangle (Easy)

Print the following pattern for n rows:

```
*****
****
***
**
*
```
(Example shown for n=5)

**Hint:** Outer loop for rows. Inner loop prints stars from 1 to (n-i+1), where i is the row number.

````python
print("Print the pattern:")
my_number = int(input("Enter a number: "))
for i in range(1, my_number+1):
    for j in range(my_number-i+1):
        print("*", end="")
    print("")
````

### Problem 4.3: Pattern Printing - Pyramid (Medium)

Print the following pyramid pattern for n rows:

```
    *
   ***
  *****
 *******
*********
```
(Example shown for n=5)

**Hint:** For row i: print (n-i) spaces, then print (2*i-1) stars. Use nested loops or string multiplication.

````python
print("Print the pattern:")
my_number = int(input("Enter a number: "))
for i in range(1, my_number+1):
    for j in range(my_number-i):
        print(" ", end="")
    for j in range(2*i-1):
        print("*", end="")
    print("")
````
