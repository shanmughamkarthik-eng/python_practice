1.	Given a string, remove leading and trailing spaces.
2.	Check whether a string starts with a given character.
3.	Find the index of the first occurrence of a character in a string.
4.	Count how many times a specific character appears in a string.
5.	Split a sentence into a list of words.
6.	Join a list of words into a single string with spaces.
7.	Replace a word in a sentence with another word

```python
string =" Learning Python is simple and Python is powerful "
string
```

```python
# 1. Given a string, remove leading and trailing spaces.

stripped_string = string.strip()
print(stripped_string)
```

```python
# 2. Check whether a string starts with a given character.

given_char = input("Enter a character: ")

if string.startswith(given_char):
    print(f"The string starts with '{given_char}'.")
else:
    print(f"The string does not start with '{given_char}'.")
```

```python
# 3. Find the index of the first occurrence of a character in a string.

char = input("Enter a character: ")
index = string.find(char)
print("The index of the first occurence of", char, "is", index)
```

```python
# 4. Count how many times a specific character appears in a string.

char = input("Enter a character: ")
count = string.count(char)
print("The character", char, "appears", count, "times in the string.")
```

```python
# 5. Split a sentence into a list of words.

words = string.split()
print(words)
```

```python
# 6. Join a list of words into a single string with spaces.

joined_string = ' '.join(words)
print(joined_string)
```

```python
# 7. Replace a word in a sentence with another word

old_word = input("Enter the word to be replaced: ")
new_word = input("Enter the new word: ")

new_string = string.replace(old_word, new_word)
print("The new string is:", new_string)
```
