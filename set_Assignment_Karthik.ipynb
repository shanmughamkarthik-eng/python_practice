```python
# Question 8: Create a set of vowels from string "hello world"
print("\nQuestion 8: Create a set of vowels from string 'hello world'")
# Your code here
string = "hello world"
vowels = {'a', 'e', 'i', 'o', 'u'}
set_vowels = set()
for char in string:
    if char in vowels:
        set_vowels.add(char)
print(f"\nThe set of vowels in the string is: {set_vowels}")
```

```python
# Question 9: Remove duplicates from list [1, 2, 2, 3, 4, 4, 5, 6, 6, 7] using set
print("\nQuestion 9: Remove duplicates from list [1, 2, 2, 3, 4, 4, 5, 6, 6, 7] using set")
# Your code here
numbers_list = [1, 2, 2, 3, 4, 4, 5, 6, 6, 7]
numbers_set = set(numbers_list)
print(numbers_set)
numbers_list = list(numbers_set)
print(numbers_list)
```

```python
# Question 10: Check if {1, 2, 3} is a subset of {1, 2, 3, 4, 5, 6}
print("\nQuestion 10: Check if {1, 2, 3} is a subset of {1, 2, 3, 4, 5, 6}")
# Your code here
numbers_set = {1, 2, 3, 4, 5, 6}
numbers_subset = {1, 2, 3}
is_subset = set(numbers_subset).issubset(set(numbers_set))
print(is_subset)
```
