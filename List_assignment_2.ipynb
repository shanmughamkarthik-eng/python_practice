```python
# Question 26: Find the person with maximum age in list of dictionaries
print("\nQuestion 26: Find the person with maximum age in list of dictionaries")
# Your code here
my_dict_list = [{'name': 'Alice', 'age': 25}, {'name': 'Bob', 'age': 30}]
max_age = 0
max_age_person = None

for my_dict in my_dict_list:
    if my_dict['age'] > max_age:
        max_age = my_dict['age']
        max_age_person = my_dict

print(f"Person with maximum age : {max_age_person}")
```

```python
# Question 30: Find the union of all sets in list of sets
print("\nQuestion 30: Find the union of all sets in list of sets")
# Your code here
my_list = [{1, 2, 3}, {4, 5, 6}, {7, 8, 9}]
union_set = set()
for my_set in my_list:
    union_set = union_set.union(my_set)
print(f"List of sets: {my_list}")
print(f"Union of all sets in list of sets: {union_set}")
```
