### 1. Count how many vowels are in a given string.

```python
print("\nCount how many vowels are in a given string")
my_string = str(input("Enter a string: "))
vowels_dict = {'a':0 , 'e':0, 'i':0, 'o':0, 'u':0}
for letter in my_string:
    if letter.lower() in vowels_dict.keys():
        vowels_dict[letter.lower()] += 1

print(vowels_dict)
```

Example: Input → 12345, Output → 54321

```python
print ("\nReverse a number using only a for loop.")
my_number = int(input("Enter a number: "))
length = len(str(my_number))
temp = my_number
reverse = 0
for i in range(length):
    reverse = (reverse * 10) + (temp%10)
    temp = temp // 10
print(f"The number {my_number } reversed is {reverse}")
```

### 3. Find the sum of digits of a number.
Example: 5483 → 20

```python
print ("\nFind the sum of digits of a number.")

my_number = int(input("Enter a number: "))
temp = my_number
sum_digits = 0

while temp > 0:
    sum_digits += temp % 10
    temp //= 10

print(f"Sum of digits of the number {my_number} is: {sum_digits}")
```

### 5. Find the factorial of a number.

```python
print ("\nFind the factorial of a number.")

my_number = int(input("Enter a number: "))
factorial = 1
for number in range (1, my_number + 1):
      factorial *= number
print(f"factorial of number {my_number} is : {factorial}")
```

### 6. Check whether a number is prime using a for loop.

```python
print("Check whether a number is prime using a for loop")
my_number = int(input("Enter a number: "))
if my_number <= 1:
    print(f"{my_number} is not a prime number")
elif my_number == 2:
    print(f"{my_number} is a prime number")
else:
    for i in range(2, my_number):
      if (my_number % i) == 0:
        print(f"{my_number} is not a prime number")
        break
    else:
        print(f"{my_number} is a prime number")
```

### 9. Find the largest number in a list. Example: [4, 7, 1, 9, 3]

```python
print("Find the largest number in a list. Example: [4, 7, 1, 9, 3]")
my_list = [4, 7, 1, 9, 3]
largest = my_list[0]
for number in my_list:
  if number > largest:
    largest = number
print(f"Largest number in the list is {largest}")
```

### 11. From 1 to 50, find:
#### a) sum of even numbers
#### b) sum of odd numbers


```python
print("From 1 to 50, find:")
print("a) sum of even numbers")
print("b) sum of odd numbers\n")

sum_even = 0
sum_odd = 0
for number in range(1, 51):
  if number % 2 == 0:
    sum_even += number
  else:
    sum_odd += number
print(f"Sum of even numbers is: {sum_even}")
print(f"Sum of odd numbers is: {sum_odd}")
```

### 12. Remove all spaces from a string.
Example: "data science rocks" → "datasciencerocks"

```python
print("Remove all spaces from a string.")
my_string = str(input("Enter a string: \n"))
new_string = ""
for letter in my_string:
  if letter == " ":
    continue
  else:
    new_string += letter
print(f"New string after removing space is:\n{new_string}")
```

### 13. Check whether a string is a palindrome.
Example: "madam"

```python
print("Check whether a string is a palindrome")
user_string = input("Enter a string: ")
length = len(user_string)
is_palindrome = True
for i in range(0, length // 2):
    if user_string[i] != user_string[length - 1 - i]:
        is_palindrome = False
        break

if is_palindrome:
    print(f"'{user_string}' is a palindrome")
else:
    print(f"'{user_string}' is not a palindrome")
```

### 14. Find duplicate values in a list.
Example: [1,2,3,2,4,5,1]


```python
print("Find duplicate values in a list.")
my_list = [1,2,3,2,4,5,1]
my_dict = {}
for number in my_list:
  if number in my_dict.keys():
        my_dict[number] += 1
        print(f"Number {number} is a duplicate")
  else:
        my_dict[number] = 1
```

### 15. Print the frequency of each character in a string.
Example: "banana"

```python
print("Print the frequency of each character in a string.")

my_string = str(input("Enter a string: "))
my_dict = {}

for letter in my_string.lower():
    if letter in my_dict.keys():
        my_dict[letter] += 1
    else:
        my_dict[letter] = 1

print(my_dict)

```
