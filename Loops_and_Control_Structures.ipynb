### Problem 1.1: Grade Calculator (Easy)
Write a program that takes a student's marks (0-100) as input and prints their grade according to the following criteria:

#### 90-100: Grade A
#### 80-89: Grade B
#### 70-79: Grade C
#### 60-69: Grade D
#### Below 60: Grade F
#### Also, validate that the marks are within the valid range (0-100).

Hint: Use if-elif-else statements. First check if marks are in valid range (0-100), then use >= comparisons starting from highest grade.

````python
print("Grade Calculator")

marks = int(input("Enter your marks: "))

if marks < 0 or marks > 100:
  print("Invalid Marks")

else:
    if marks>= 90 and marks<=100:
      print("Grade A")

    elif marks >= 80 and marks <= 89:
      print("Grade B")

    elif marks >= 70 and marks <= 79:
      print("Grade C")

    elif marks >= 60 and marks <= 69:
      print("Grade D")

    else:
      print("Grade F")
````

### Problem 1.2: Leap Year Checker (Medium)
Write a program to check if a given year is a leap year or not.

###Rules:

#####A year is a leap year if it is divisible by 4
#####However, if it is divisible by 100, it is NOT a leap year
#####Exception: If it is also divisible by 400, then it IS a leap year

###Examples:

#####2000: Leap year (divisible by 400)
#####1900: Not a leap year (divisible by 100 but not 400)
#####2024: Leap year (divisible by 4 but not 100)
#####Hint: Check conditions in order: first divisible by 400, then by 100, then by 4. Use modulo operator (%) to check divisibility.

````python
print("Write a program to check if a given year is a leap year or not.")

year = int(input("Please enter a year: "))
if (year % 4 == 0 and year % 100 != 0) or (year % 400 == 0):
    print(f"{year} is a leap year.")
else:
    print(f"{year} is not a leap year.")
````

### Problem 1.3: Triangle Type Identifier

##### Write a program that takes three sides of a triangle as input and determines:

##### Whether the three sides can form a valid triangle (sum of any two sides must be greater than the third side)

##### If valid, classify the triangle as:
##### Equilateral (all three sides equal)
##### Isosceles (exactly two sides equal)
##### Scalene (no sides equal)

##### Hint: First check: (a+b>c) and (b+c>a) and (a+c>b). Then use == to compare sides: if all three equal, it's equilateral; if any two equal, it's isosceles; otherwise scalene.

````python
print("Write a program that takes three sides of a triangle as input and determines:")

side_a = float(input("Enter the first side of the triangle: "))
side_b = float(input("Enter the second side of the triangle: "))
side_c = float(input("Enter the third side of the triangle: "))

if (side_a + side_b > side_c) and (side_b + side_c > side_a) and (side_a + side_c > side_b):
      print("The triangle is valid.")

      if side_a == side_b == side_c:
          print("The triangle is equilateral.")
      elif side_a == side_b or side_b == side_c or side_a == side_c:
          print("The triangle is isosceles.")
      else:
          print("The triangle is scalene.")
else:
  print("The triangle is not valid.")
````

### Problem 1.4: BMI Calculator with Categories (Easy)

Write a program that calculates BMI (Body Mass Index) and categorizes it.

Formula: BMI = weight (kg) / (height (m))²

Categories:
- Below 18.5: Underweight
- 18.5-24.9: Normal weight
- 25-29.9: Overweight
- 30 and above: Obese

**Hint:** First calculate BMI using the formula (use ** for power or multiply height by itself). Then use if-elif-else to categorize.

````python
print("Write a program that calculates BMI (Body Mass Index) and categorizes it.")

weight = float(input("Enter your weight in kg: "))
height = float(input("Enter your height in meters: "))

body_mass_index = weight / (height ** 2)

if body_mass_index < 18.5:
    print("Under weight")
elif body_mass_index < 24.9:
    print("Normal weight")
elif body_mass_index < 29.9:
    print("Over weight")
else:
    print("Obese")
````

### Problem 2.2: Factorial Calculator (Medium)

Write a program to calculate the factorial of a number using a for loop.

Factorial of n (n!) = n × (n-1) × (n-2) × ... × 1

Example: 5! = 5 × 4 × 3 × 2 × 1 = 120

**Hint:** Initialize result = 1. Use for loop with range(1, n+1) and multiply result by each number.

````python
print ("\nFind the factorial of a number.")

my_number = int(input("Enter a number: "))
factorial = 1
for number in range (1, my_number + 1):
      factorial *= number
print(f"factorial of number {my_number} is : {factorial}")
````

### Problem 2.3: Sum of Even Numbers (Easy)

Write a program that calculates the sum of all even numbers between 1 and n (inclusive), where n is provided by the user.

Example: If n = 10, sum = 2 + 4 + 6 + 8 + 10 = 30

**Hint:** Use for loop with range(1, n+1). Check if i % 2 == 0, then add to sum. Or use range(2, n+1, 2) to iterate only even numbers.

````python
print("Write a program that calculates the sum of all even numbers between 1 and n")
n = int(input("Enter a number: "))
sum_even = 0

for number in range(2, n+1, 2):
    sum_even += number
print(f"Sum of even numbers is: {sum_even}")
````

Write a program to print the first n terms of the Fibonacci series.

The Fibonacci series: 0, 1, 1, 2, 3, 5, 8, 13, 21, ...

Each number is the sum of the two preceding ones.

**Hint:** Initialize a=0, b=1. Print a, then in loop: calculate next=a+b, print next, then update a=b and b=next.

````python
print("Write a program to print the first n terms of the Fibonacci series.")

n = int(input("Enter a number: "))

a = 0
b = 1
print(f"The first '{n}' terms of the Fibonacci series:")
for number in range(0, n):
    print(a)
    next = a + b
    a = b
    b = next
````

### Problem 2.5: Prime Number Checker (Medium)

Write a program to check if a given number is prime or not.

A prime number is a natural number greater than 1 that has no positive divisors other than 1 and itself.

**Hint:** Check if n is divisible by any number from 2 to n-1. If no divisor found, it's prime. Optimize: check only up to √n.

````python
print("Write a program to check if a given number is prime or not.")

my_number = int(input("Enter a number: "))

if my_number <= 1:
    print(f"{my_number} is not a prime number")
elif my_number == 2:
    print(f"{my_number} is a prime number")
else:
    for i in range(2, my_number):
      if (my_number % i) == 0:
        print(f"{my_number} is not a prime number")
        break
    else:
        print(f"{my_number} is a prime number")
````

### Problem 3.1: Reverse a Number (Medium)

Write a program that reverses the digits of a given number using a while loop.

Example: Input: 12345, Output: 54321

**Hint:** Initialize reverse=0. While n>0: get last digit (n%10), add to reverse (reverse=reverse*10+digit), remove last digit (n=n//10).

````python
print("Write a program that reverses the digits of a given number using a while loop.")
my_number = int(input("Enter a number: "))
length = len(str(my_number))
temp = my_number
reverse = 0
while temp > 0:
    reverse = (reverse * 10) + (temp%10)
    temp = temp // 10
print(f"The number {my_number } reversed is {reverse}")
````

### Problem 3.2: Sum of Digits (Easy)

Write a program that calculates the sum of all digits in a given number.

Example: Input: 12345, Output: 15 (1+2+3+4+5)

**Hint:** While n>0: add last digit to sum (sum += n%10), then remove it (n = n//10).

````python
print("Write a program that calculates the sum of all digits in a given number.")
my_number = int(input("Enter a number: "))
temp = my_number
sum_digits = 0

while temp > 0:
    sum_digits += temp % 10
    temp //= 10

print(f"Sum of digits of the number {my_number} is: {sum_digits}")
````
